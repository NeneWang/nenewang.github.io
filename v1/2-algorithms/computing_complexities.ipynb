```python
import pprint

def can_partition(S):
    total_sum = sum(S)
    
    # If the total sum is odd, we cannot partition it into two equal subsets
    if total_sum % 2 != 0:
        return False
    
    target_sum = total_sum // 2
    n = len(S)
    
    # Initialize a DP array
    dp = [False] * (target_sum + 1)
    dp[0] = True  # Base case: A subset with sum 0 is always possible (empty subset)
    
    # Fill the DP array
    for num in S:
        for i in range(target_sum, num - 1, -1):
            dp[i] = dp[i] or dp[i - num]
    print(dp)
    return dp[target_sum]

# Example usage
S = [1, 5, 11, 5]
print(can_partition(S))  # Output: True

S = [1, 2, 3, 5]
print(can_partition(S))  # Output: False

S = [5, 13, 8, 17, 21, 23]
print(can_partition(S))  # Output: False


S = [5, 13, 8, 17, 20, 23]
print(can_partition(S))
```

```python
def subset_sum(numbers, target):
    num_count = len(numbers)
    
    # Create a 2D list (num_count + 1) x (target + 1) initialized to False
    dp = [[False] * (target + 1) for _ in range(num_count + 1)]
    
    # A subset with sum 0 is always possible (empty subset)
    for i in range(num_count + 1):
        dp[i][0] = True  
    
    # Fill the dp table
    for i in range(1, num_count + 1):
        for current_sum in range(1, target + 1):
            if numbers[i-1] > current_sum:
                dp[i][current_sum] = dp[i-1][current_sum]
            else:
                dp[i][current_sum] = dp[i-1][current_sum] or dp[i-1][current_sum - numbers[i-1]]
    
    return dp[num_count][target]

# Example usage
numbers = [3, 34, 4, 12, 5, 2]
target = 9
print(subset_sum(numbers, target))  # Output: True
```

```python
# Define the formula as a function
def formula(p, q, r):
    # Using logical expressions
    return (p or not q) and (not q or r) and (not r or p)

# Generate all possible combinations of truth values for p, q, r
def is_satisfiable():
    # Truth values: True and False
    truth_values = [True, False]
    
    # Iterate through all combinations of p, q, and r
    for p in truth_values:
        for q in truth_values:
            for r in truth_values:
                # Evaluate the formula for the current combination
                if formula(p, q, r):
                    # If the formula is true for any combination, it is satisfiable
                    print(f"Satisfiable with p={p}, q={q}, r={r}")
                    return True
    
    # If no combination satisfies the formula, it is unsatisfiable
    print("Unsatisfiable")
    return False

# Run the satisfiability check
is_satisfiable()
```
