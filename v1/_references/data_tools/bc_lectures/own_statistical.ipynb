```python
# Wine testings.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
```

```python
# df = pd.read_csv('https://github.com/dsrscientist/DSData/blob/master/winequality-red.csv')
df = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/DSData/master/winequality-red.csv')
df.head()
```

```python
# Correlation using sns replot residual sugar and density

# Relationship. This is the mean squared error relationship . Any others would be worst than this. This is the BEST approximation you COULD have made.
# Answers the question: How much does an increase of 1 gram of 1 residual sugar corresponds to in density?
import seaborn as sns
sns.regplot(x='residual sugar', y='density', data=df, line_kws={'color': 'black'})
plt.show()
```
