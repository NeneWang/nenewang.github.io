Differences (1 points)
Create a new boolean column in the DataFrame which is True if the camera was released on or after
2004, and False if it was released prior to 2004.
On average, is the mean weight of cameras different across these two time periods? Is the difference
significant?
On average, is the mean price different across these two time periods? Is the difference significant?





```python
# Create a boolean difference before and after 2004
cameras_before_2004 = cameras[cameras['Release date'] < 2004]
cameras_after_2004 = cameras[cameras['Release date'] >= 2004]
# find mean and compare.
before_price_mean = cameras_before_2004['Price'].mean()
after_price_mean = cameras_after_2004['Price'].mean()

before_weight_mean = cameras_before_2004['Weight (inc. batteries)'].mean()
after_weight_mean = cameras_after_2004['Weight (inc. batteries)'].mean()

print(f'Price mean before 2004: {before_price_mean}')
print(f'Price mean after 2004: {after_price_mean}')
print(f'Weight mean before 2004: {before_weight_mean}')
print(f'Weight mean after 2004: {after_weight_mean}')

```

```python
def plot_scatter_mean_year(cameras, year=2004, x_column='Price', y_column='Weight (inc. batteries)', x_color='red', y_color='blue', show_mean=True):

    # Create a boolean difference before and after 2004
    cameras_before_2004 = cameras[cameras['Release date'] < year]
    cameras_after_2004 = cameras[cameras['Release date'] >= year]
    # find mean and compare.
    before_price_mean = cameras_before_2004[x_column].mean()
    after_price_mean = cameras_after_2004[x_column].mean()

    before_weight_mean = cameras_before_2004[y_column].mean()
    after_weight_mean = cameras_after_2004[y_column].mean()

    # scatter plot red and blue, before and after.

    plt.scatter(cameras_before_2004[x_column], cameras_before_2004[y_column], color='red', label='Before 2004', alpha=0.4)
    plt.scatter(cameras_after_2004[x_column], cameras_after_2004[y_column], color='blue', label='After 2004', alpha=0.4)

    # blue y line and red y line indicating mean weight.
    plt.axhline(before_price_mean, color=x_color, linestyle='--', label='Mean Price Before 2004')
    plt.axhline(after_price_mean, color=y_color, linestyle='--', label='Mean Price After 2004')

    plt.axvline(before_weight_mean, color=x_color, linestyle='--', label='Mean Weight Before 2004')
    plt.axvline(after_weight_mean, color=y_color, linestyle='--', label='Mean Weight After 2004')
    plt.title(f'Scatter plot of {x_column} vs {y_column} before and after {year}')
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    
    # disable legends if show_mean is False
    if not show_mean:
        plt.legend().remove()
    return {'before_price_mean': before_price_mean, 'after_price_mean': after_price_mean, 'before_weight_mean': before_weight_mean, 'after_weight_mean': after_weight_mean, 
            'count_before': len(cameras_before_2004), 'count_after': len(cameras_after_2004),
            'percentage_count_difference': (len(cameras_after_2004) - len(cameras_before_2004)) / len(cameras_before_2004) * 100
        }

data = plot_scatter_mean_year(cameras, 2004, 'Weight (inc. batteries)', 'Price', 'red', 'blue')
plt.legend()
plt.show()
import pprint
pprint.pprint(data)
```

- It seems that there is a more signigicant difference in weight before and after 2004.
- While there is a difference in price, it is not as significant.

```python
# Plot bar chart count of cameras before and after 2004
plt.bar(['Before 2004', 'After 2004'], [cameras_before_2004.shape[0], cameras_after_2004.shape[0]])
```

Do all camera brands follow this general trend, or do some differ? In other words, does the brand of
camera have an effect on the number of cameras released pre- or post-2004? If there are differences
among brands, which brands were closest to the general trends in pre-2004/post-2004 camera releases,
and which brands differed the most?



```python
# It would be interesting now to plot together, a scatter plot with the mean of each brand
unique_brands = cameras['Brand'].unique()
mean_df_arr = []
all_colors = ['red', 'blue', 'green', 'yellow', 'black', 'pink', 'purple', 'orange', 'brown', 'gray', 'cyan', 'magenta', 'olive', 'maroon', 'navy', 'teal', 'aqua', 'lime', 'fuchsia', 'silver', 'white']
for idx, brand in enumerate(unique_brands):
    
    camera_brand = cameras[cameras['Brand'] == brand]
    color_1 = all_colors[idx%len(all_colors)] 
    mean_data = plot_scatter_mean_year(camera_brand, 2004, 'Weight (inc. batteries)', 'Price', color_1, color_1, show_mean=False )
    mean_data['Brand'] = brand
    mean_data['price_diff'] = abs(mean_data['before_price_mean'] - mean_data['after_price_mean'])
    mean_data['weight_diff'] = abs(mean_data['before_weight_mean'] - mean_data['after_weight_mean'])
    mean_data['price_trend_diff'] = mean_data['before_price_mean'] - mean_data['after_price_mean']
    mean_data['weight_trend_diff'] = mean_data['before_weight_mean'] - mean_data['after_weight_mean']
    
    
    mean_df_arr.append(mean_data)
    # plt.legend()
plt.show()
mean_df = pd.DataFrame(mean_df_arr)
    
    
    
    
```
