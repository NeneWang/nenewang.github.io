```python
import random

# Simulating heights of a population with a mean of 170 cm and a standard deviation of 10 cm
height = random.gauss(170, 10)
print("Simulated height:", height)
```

```python
import random

# Simulating data for statistical analysis
data = [random.gauss(0, 1) for _ in range(1000)]
# Chart data.
import matplotlib.pyplot as plt
plt.hist(data, bins=30)
```

```python
import random

# Simulating a random walk with Gaussian steps
position = 0
steps = 1000
position_at_step = []
for _ in range(steps):
    position += random.gauss(0, 1)
    position_at_step.append(position)
    # print(position, position_at_step[-1])

# Chart.
plt.plot(position_at_step)
```

```python
import random

# Simulating stock price movements with a mean return of 0.05 and a standard deviation of 0.2
initial_price = 100
drift = 0.05
volatility = 0.2
days = 30
price_at = []

for _ in range(days):
    price_change = random.gauss(drift, volatility)
    initial_price *= 1 + price_change
    price_at.append(initial_price)

# Chart.
plt.plot(price_at)
```
