```

2 | 3 | 4


  | c 
--|--
```

```py

```

```
  | a | b | c
--|--|--|--
0 | 1 | 2 | 5
1 | 2 | 3 | 6
2 | 3 | 4 | nan
6 | nan | nan | 7
```



`````python
# Demostration:

import pandas as pd
df1 = pd.DataFrame({'a':[1, 2, 3], 'b': [2, 3, 4]}, index=[0, 1, 2])
df2 = pd.DataFrame({'c': [5, 6, 7]}, index=[0, 1, 6])


# Concat:
pd.concat([df1, df2], axis=1)
`````
