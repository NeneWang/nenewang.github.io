```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Wine data df
df = pd.read_csv('https://gist.githubusercontent.com/tijptjik/9408623/raw/b237fa5848349a14a14e5d4107dc7897c21951f5/wine.csv', delimiter=',')
df.head()
```

```python
# Show columns
df.columns
```

```python
# Correlations of wine.
correlations = df.corr()

# Plot matrix using sns
import seaborn as sns
sns.heatmap(correlations, annot=True)
plt.show()
```

```python
# Scatter plot density and residual sugar.
plt.scatter(df['Flavanoids'], df['Nonflavanoid.phenols'])
plt.xlabel('Density')
plt.ylabel('Residual Sugar')
plt.show()
```
