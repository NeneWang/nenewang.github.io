# 3 - Simple Plot Example

This notebook shows a basic plot using matplotlib's pyplot.

```python
# Import matplotlib
import matplotlib.pyplot as plt
```

```python
# Create some simple data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
```

```python
# Create a plot
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set(title='Simple Line Plot', xlabel='x-axis', ylabel='y-axis')
plt.show()
```
