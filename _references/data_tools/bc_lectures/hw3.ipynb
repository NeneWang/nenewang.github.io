## Pandas Questions (12 points)
1. What is the difference between a Series and a DataFrame?

> A Pandas Series is a one-dimensional array of indexed data.
> Dataframe is an analog of two-dimensional array. Could be thinked as a sequence of aligned (shares same index.) series objects. A dataframe would be holding multiple Series objects.

Definitions taken from [3.01](https://jakevdp.github.io/PythonDataScienceHandbook/03.01-introducing-pandas-objects.html)

2. What role do indexes play in both Series and DataFrame objects?

> Indexes play the role of unique labels to access to rows data. Similar to how in a dictionary, keys are used to reference values inside of a dict.


3. In your own words, describe the difference between the loc and iloc attributes available on Series
objects.

> `loc` is used to index and slice (select range) referencing the explicit index. Which could be thinked as the 'label' that you provide to refer a specific row.

> `iloc` is used to index and slice referencing the Implicit (You can think of it as array position) inside of an array.


4. In your own words, describe the difference between the `loc` and `iloc` attributes available on DataFrame objects.

Similar to how the selection of `loc` and `iloc` would be used to select and slice in a series, however now we have to consider the option of selecting an extra dimension

Suppose you have a dataframe:

User | Salary | Education
--- | -- | --
a | 1000 | Highschool
b | 2000 | college
c | 3000 | phd


> `loc` used to select using the explicit index and column names/labels e.g.

You can select using the explicit column and index name:
```py
df.loc['b':, 'Salary':]
```

And the expected return would be:

Salary | Education
--|--
2000 | College
3000 | phd

> `iloc` used to select using the index as an array to ubicate the row of data and Column data to be selected

```py
df.iloc[:2]
```
User | Salary | Education
--- | -- | --
a | 1000 | Highschool
b | 2000 | college

Both of them can be used to select in the 2 dimension or 1 of them:

```py
df.iloc[1:3, 1:3]
```

5. In what way is a Series like a dictionary? In what way is it like a list or one-dimensional array?

> Series is like a dictionary in the following ways:
- collection of values through the equivalent of a key, which is the index.
- A series can be modified in a dictionary like syntax replacing it's value, by referencing the index
- You can also extend the series by assigning values into a new index value.

> Ways it is similar to an one-dimensional array
-  operations such as: Slicing, and Masking. and range indexing
