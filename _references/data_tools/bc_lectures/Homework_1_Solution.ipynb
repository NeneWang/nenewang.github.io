## Vectorized Matrix Addition Benchmarking

The testing procedure in this notebook shows the results for matrix_add_2d with lists, np.add() with lists, and np.add() with arrays.

```python
def matrix_add_2d(a, b):
  output_arr = []

  for row_a, row_b in zip(a, b):
    output_row = []
    for i, j in zip(row_a, row_b):
      output_row.append(i+j)
    output_arr.append(output_row)

  return output_arr
```

```python
# Small

a_small = [[random.randint(-10, 10) for _ in range(100)] for _ in range(100)]
b_small = [[random.randint(-10, 10) for _ in range(100)] for _ in range(100)]

a_small_arr = np.array(a_small)
b_small_arr = np.array(b_small)

print("matrix_add_2d() with lists")
%timeit matrix_add_2d(a_small, b_small)
print("np.add() with lists")
%timeit np.add(a_small, b_small)
print("np.add() with arrays")
%timeit np.add(a_small_arr, b_small_arr)
```

```python
# Medium

a_medium = [[random.randint(-10, 10) for _ in range(1000)] for _ in range(1000)]
b_medium = [[random.randint(-10, 10) for _ in range(1000)] for _ in range(1000)]

a_medium_arr = np.array(a_medium)
b_medium_arr = np.array(b_medium)

print("matrix_add_2d() with lists")
%timeit matrix_add_2d(a_medium, b_medium)
print("np.add() with lists")
%timeit np.add(a_medium, b_medium)
print("np.add() with arrays")
%timeit np.add(a_medium_arr, b_medium_arr)
```

```python
# Large

a_large = [[random.randint(-10, 10) for _ in range(10000)] for _ in range(10000)]
b_large = [[random.randint(-10, 10) for _ in range(10000)] for _ in range(10000)]

a_large_arr = np.array(a_large)
b_large_arr = np.array(b_large)

print("matrix_add_2d() with lists")
%timeit matrix_add_2d(a_large, b_large)
print("np.add() with lists")
%timeit np.add(a_large, b_large)
print("np.add() with arrays")
%timeit np.add(a_large_arr, b_large_arr)
```

## Written questions

1. There are many reasons to prefer NumPy over a handwritten function. NumPy is already written and includes a useful variety of math functions, which means you do not have to write them all yourself. Rewriting all the math operations you need from scratch is time-consuming and opens the possibility of introducing bugs, which NumPy already guards against with an extensive test suite. A significant amount of work was done to integrate NumPy with Python language features, so it is possible to write very simple and compact code without additional effort.

2. Giving NumPy arrays is much faster than giving lists. The reason is that NumPy is written to work with arrays, not native Python objects. Any lists given as arguments are converted to arrays internally before doing any work. The time difference you see in the cells above is because, when given lists, NumPy has to repeatedly convert the lists to arrays every time the function is called.
